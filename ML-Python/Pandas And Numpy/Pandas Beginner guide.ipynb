### *Performing Arithmetic Operations - Crunching Numbers*
Our next adventure involves performing arithmetic operations. Suppose we want to understand how doubling the price         impacts our dataset. Let’s create a new column for this.

```python
# Creating a new column with doubled Price
data['Double_Price'] = data['Price']-data['Price'] * (20/100)

# Display the first few rows to see the new column
print(data[['Price', 'Double_Price']].head())
```

```python
0="row_wise"
1="column wise"
```

```python
def calculate_discount_price(data):
    
    rating = data['Rating']
    price = data['Price']
    # Apply the discount logic based on the rating
    if rating < 4.0:
        # Apply a 30% discount
        discounted_price = price * 0.7
    elif rating > 4.5:
        # Double the price
        discounted_price = price * 2
    else:
        # No change in price
        discounted_price = price
    
    return discounted_price
```

```python
data['discounted_price'] = data.apply(calculate_discount_price, axis=1)
```

### *Data Aggregation - Summarizing Data*
To gain quick insights, we can summarize our data through aggregation. For instance, let’s calculate the average price     of phones grouped by their company.

```python
# Aggregating data: Calculating the mean Price by company
mean_price_by_company = data.groupby(['company','Android_version'])[['Price','Spec_score']].sum()

# # Display the aggregated data
# print(mean_price_by_company)
mean_price_by_company
```

```python
data.groupby('Android_version')
```

```python
# Group by 'mobile' and calculate the sum of 'units_sold' and average 'price'
data.groupby('company').agg({
    'Spec_score': 'sum',  # Sum of units_sold
    'Price': 'mean',
    'Rating':'median'
    # Average price
}).sort_va
```
